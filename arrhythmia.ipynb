## Visual exploration of relationship between variables
For Sex Vs Heart Rate we shall use a Box Plot

If you have a categorical independent variable and a continuous target variable, you can use a one-way ANOVA (Analysis of Variance) test to determine whether there is a significant difference between the means of the target variable across the categories of the independent variable.

```python
from scipy.stats import f_oneway

# Conduct one-way ANOVA test
group1 = df[df['Sex'] == 'Male']['Heart rate']
group2 = df[df['Sex'] == 'Female']['Heart rate']
f_statistic, p_value = f_oneway(group1, group2)

# Print results
print('F statistic:', f_statistic)
print('p-value:', p_value)
```

# Interpretation
The results from the ANOVA test comparing sex and heart rate show an F-statistic of 1.7736 and a p-value of 0.1836.

The F-statistic measures the ratio of the variance between the two groups (male and female) to the variance within the groups. The higher the F-statistic, the more likely it is that there is a significant difference between the groups. In this case, the F-statistic is 1.7736, which is relatively small, indicating that there may not be a significant difference between male and female heart rates.

The p-value is the probability of observing a test statistic as extreme as the one calculated from the sample data, assuming that the null hypothesis (i.e., no difference between male and female heart rates) is true. In this case, the p-value is 0.1836, which is greater than the commonly used significance level of 0.05. This suggests that we cannot reject the null hypothesis and conclude that there is a significant difference between male and female heart rates. However, it is important to note that this conclusion depends on the chosen significance level and the specific context of the study.

## Regression tests for continuous data

```python
from scipy.stats import linregress
from tabulate import tabulate

variable_pairs = [('Age', 'Heart rate'),
                  ('Height', 'Heart rate'),
                  ('Weight', 'Heart rate'),
                  ('QRS Duration', 'Heart rate'),
                  ('P-R interval', 'Heart rate'),
                  ('Q-T interval', 'Heart rate'),
                  ('T interval', 'Heart rate'),
                  ('P interval', 'Heart rate'),
                  ('QRS', 'Heart rate'),
                  ('T', 'Heart rate'),
                  ('P', 'Heart rate'),
                  ('QRST', 'Heart rate'),
                   ('J', 'Heart rate')]

results = []
for var in variable_pairs:
    slope, intercept, r_value, p_value, std_err = linregress(df[var[0]], df[var[1]])
    results.append([var[0], var[1], slope, intercept, r_value, p_value, std_err])

# print the results as a table
headers = ['Variable 1', 'Variable 2', 'Slope', 'Intercept', 'R-Value', 'P-Value', 'Std Err']
print(tabulate(results, headers=headers))
```

## Overall Interpretation
From the table above, Weight is the only variable with a P-Value below 0.05. The detailed interpretation of the Weight variable results are given below

## Weight linear regression test results interpretation

The slope of the regression line is -0.1461, which means that for every one-unit increase in weight, the heart rate will decrease by 0.1461 units. In other words, there is a negative relationship between weight and heart rate, where higher weight is associated with lower heart rate.

The intercept of the regression line is 84.4179, which represents the predicted value of the heart rate when the weight is zero. However, since weight cannot be zero in practice, the intercept is not very meaningful in this context.

The R-squared value of 0.0306 indicates that only about 3% of the variation in heart rate can be explained by weight in this dataset. This suggests that weight is not a strong predictor of heart rate.

The p-value of 0.0002 indicates that the slope of the regression line is statistically significant at a significance level of 0.05. This means that we can reject the null hypothesis that the slope is zero, and conclude that weight has a statistically significant effect on heart rate. However, the low R-squared value suggests that the effect size may be small or that there may be other variables that are more important predictors of heart rate.

## Regression tests for BMI

```python
from scipy.stats import linregress
slope, intercept, r_value, p_value, std_err = linregress(df['BMI'], df['Heart rate'])

# Print results
print('Slope:', slope)
print('Intercept:', intercept)
print('R-squared:', r_value**2)
print('p-value:', p_value)
```

## Interpretation
The regression test is examining the relationship between BMI and heart rate. The results indicate:

Slope: -0.27002919095789923
The slope indicates the amount of change in the heart rate for each unit increase in BMI. Here, for each unit increase in BMI, the heart rate decreases by -0.27002919095789923.

Intercept: 85.58985637192909
The intercept represents the predicted value of heart rate when BMI is zero. Here, the predicted heart rate is 85.58985637192909 when BMI is zero. However, since BMI cannot be zero, this value is not practically meaningful in this context.

R-squared: 0.03265066633152751
R-squared indicates the proportion of variance in the heart rate that is explained by the BMI. Here, the BMI explains only 3.27% of the variance in heart rate.

p-value: 0.00011406954528936724
The p-value indicates the statistical significance of the relationship between BMI and heart rate. Here, the p-value is less than 0.05, which means that the relationship is statistically significant. In other words, it is highly unlikely to observe such an association if there was no real association in the population.

Overall, the results indicate that there is a statistically significant negative association between BMI and heart rate. However, the effect size is relatively small, with BMI explaining only a small proportion of the variance in heart rate. It is also important to note that correlation does not imply causation, and other factors may be influencing the relationship between BMI and heart rate.
